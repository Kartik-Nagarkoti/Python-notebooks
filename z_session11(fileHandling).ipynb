# Context Manager [with]

```python
#it aotomatically closes the file
```

```python
with open("sample4.txt","w") as f:
    f.write("Happy Birthday")
    f.write("Happy Birthday")
#f.write("Happy Birthday")   this will not execute cause "with" has closed file
```

```python
# suppose pc capacity is : 8GB
#file size               : 32GB

#in such case we will load file in chunks
```

```python
#suppose a big file
big=[]
for i in range(200):
    big.append(f"bigNumber:{i}")

with open("sample5.txt","w") as f:
    f.writelines(big)
```

```python
chunk_size = 50
with open("sample5.txt", "r") as f:
    while True:
        chunk = f.read(chunk_size)
        if not chunk:
            break
        print(chunk)
        print("\n")
```

```python
#Replace particular text:
#step 1: open file in read mode and save content in a variable (c)
#step 2: use replace keyword to change content and save it in another variable (u=c.replace("hello","hi")
#srep 3: open file in write mode and feed latest vriable (f.write(u))
```

```python
with open("sample 3.txt","r") as f:
    content=f.read()

update=content.replace("hello","hi is that you")

with open("sample 3.txt","w") as f:
    f.write(update)
```

```python
#tell and seek

#tell() : used to know current pointer location
#seek() : helps in relocating pointer

with open("sample 3.txt","r") as f:
    print(f.read(10))
    print(f.tell())
    print(f.read(5))
    print(f.tell())
    f.seek(2)
    print(f.tell())
    print(f.read(5))
```
